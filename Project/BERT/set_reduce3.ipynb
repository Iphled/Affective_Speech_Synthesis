```python
import pandas as pd
from transformers import pipeline

df = pd.read_csv("data/BERT_annotated_data.csv")
df
```

```python
falsche_zeilen = df.loc[df["emotions"] != df["emotion_label"]]
len(falsche_zeilen) / len(df["emotions"])
```

```python

surprise_zeilen = df.loc[df["emotion_label"] == "surprise"]
len(surprise_zeilen) / len(df["emotion_label"])
print("Fear: " + str(len(surprise_zeilen.loc[df["emotions"] == "fear"]) / len(surprise_zeilen)))
print("Joy: " + str(len(surprise_zeilen.loc[df["emotions"] == "joy"]) / len(surprise_zeilen)))
print("Sadness: " + str(len(surprise_zeilen.loc[df["emotions"] == "sadness"]) / len(surprise_zeilen)))
print("Anger: " + str(len(surprise_zeilen.loc[df["emotions"] == "anger"]) / len(surprise_zeilen)))
```

```python

surprise_fear = surprise_zeilen.copy()
surprise_joy = surprise_zeilen.copy()
surprise_sad = surprise_zeilen.copy()
surprise_ang = surprise_zeilen.copy()
```

```python
len(surprise_fear)
```

```python
def change_fear(text):
    return "I am extremely worried. "+text
#surprise_fear['text'] = surprise_fear['text'].apply(change_fear)


def change_joy(text):
    return "I am so happy. "+text 
#surprise_joy['text'] = surprise_joy['text'].apply(change_joy)


def change_sad(text):
    return "I am very sad. "+text + ""
#surprise_sad['text'] = surprise_sad['text'].apply(change_sad)


def change_ang(text):
    return "I am very angry. "+text + ""

surprise_ang = surprise_ang['text'].apply(change_ang).values.tolist()
surprise_sad = surprise_sad['text'].apply(change_sad).values.tolist()
surprise_joy = surprise_joy['text'].apply(change_joy).values.tolist()
surprise_fear = surprise_fear['text'].apply(change_ang).values.tolist()
```

```python
len(surprise_joy)
```

```python

model = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")
```

```python
joy_emotions=model(surprise_joy)
joy_emotions
```

```python
ang_emotions=model(surprise_ang)
ang_emotions
```

```python
sad_emotions=model(surprise_sad)
sad_emotions
```

```python
fear_emotions=model(surprise_fear)
fear_emotions
```

```python
surprise_zeilen
```

```python
newem=list()
for i in range(len(surprise_fear)):
    l=list()
    l.append(joy_emotions[i]['label'])
    l.append(sad_emotions[i]['label'])
    l.append(fear_emotions[i]['label'])
    l.append(ang_emotions[i]['label'])
    joypos=list()
    joypos.append(joy_emotions[i]['score'])
    joypos.append(sad_emotions[i]['score'])
    joypos.append(fear_emotions[i]['score'])
    joypos.append(ang_emotions[i]['score'])
    score=0
    label='neutral';
    for i in range(len(l)):
        if l[i]!='surprise' and joypos[i]>score:
            score=joypos[i]
            label=l[i]
    newem.append(label)
        
surprise_zeilen["emotion_label"]=newem    
```

```python
surprise_zeilen.loc[:,"emotion_label"]=newem  
```
