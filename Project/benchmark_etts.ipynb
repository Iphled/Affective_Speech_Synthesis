```python
from scipy.io.wavfile import read
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np


rate,data=read("data/benchmark_audios/TTS_easy/1.wav")
length = data.shape[0] / rate
#plotting the audio time and amplitude
time = np.linspace(0., length, data.shape[0])
values=list()
for i in range(data.shape[0]):
        values.append(data[i])
plt.plot(time, values, label="Audio")
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```

```python
from scipy.io.wavfile import read
from pydub import AudioSegment
from Project.write_back_audio import overlay_emotion

emotion_values = ("Neutral", "Joy", "Sadness", "Anger", "Fear")

for i in range(12):
    rate, data = read("data/benchmark_audios/TTS_easy/"+str(i)+".wav")
    audio = AudioSegment.from_file("data/benchmark_audios/TTS_easy/"+str(i)+".wav", 'r')
    length = data.shape[0] / rate
    for emotion in emotion_values:
        audio2=overlay_emotion(audio,length,emotion)
        audio2.export("data/benchmark_audios/ETTS_easy/"+emotion+str(i)+".wav",format="wav")
```
