```python
from scipy.io.wavfile import read
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
rate,data=read("data/test.wav")
print(f"number of channels = {data}")
length = data.shape[0] / rate
print(f"length = {length}s")
#plotting the audio time and amplitude
time = np.linspace(0., length, data.shape[0])
```

```python
values=list()
for i in range(data.shape[0]):
    if data[i][0]>data[i][1]:
        values.append(data[i][0])
    else:
        values.append(data[i][1])
        
values
```

```python
plt.plot(time, values, label="Audio")
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```

```python
import librosa
y, sr = librosa.load('data/test.wav')
chromagram_stft = librosa.feature.chroma_stft(y=y, sr=sr)
fig, ax = plt.subplots(nrows=1, sharex=True, sharey=True)
librosa.display.specshow(chromagram_stft, y_axis='chroma', x_axis='time', ax=ax)
ax.set(ylabel='STFT')
ax.label_outer()
```

```python
l=list()
for i in range(chromagram_stft.shape[1]):
    max=0
    line=-1
    for j in range(chromagram_stft.shape[0]):
        if chromagram_stft[j][i]>max:
            max=chromagram_stft[j][i]
            line=j
    l.append(line)
l
```

```python
plt.plot(range(len(l)), l, label="Audio")
plt.legend()
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```
